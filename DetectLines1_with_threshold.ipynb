```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
import cv2
```

```python
image_color = mpimg.imread('11.jpg')
plt.imshow(image_color)
```

```python
image_color.shape # this is 3 coloured image
```

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_RGB2GRAY) #Corrected... image is RGB, not BGR.
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape #now this is binary image
```

```python
image_copy = np.copy(image_gray)
image_copy.shape
```

```python
image_copy[ (image_copy[:,:] < 220) ] = 0  # any value that is not white colour
```

```python
# Display the image                 
plt.imshow(image_copy, cmap = 'gray')
plt.show()
```
